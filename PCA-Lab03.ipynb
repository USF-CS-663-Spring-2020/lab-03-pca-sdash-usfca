```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('wdbc.data', header=None)
#df.head()
#df.keys()

df.shape

column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error',
       'fractal dimension error', 'worst radius', 'worst texture',
       'worst perimeter', 'worst area', 'worst smoothness',
       'worst compactness', 'worst concavity', 'worst concave points',
       'worst symmetry', 'worst fractal dimension']

```

```python
df.columns =column_labels
df

df['diagnosis'].replace('M',1,inplace=True)
df['diagnosis'].replace('B',0,inplace=True)
```

```python
feature_df = df.iloc[:,2:]
feature_df
```

```python
label_df = df.iloc[:,1]
label_df

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)

x = feature_df.values
x = StandardScaler().fit_transform(x)
np.mean(x),np.std(x)
```

```python
from sklearn.decomposition import PCA
pca  = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data);
scaled_data.shape
```

```python
x_pca.shape

plt.figure(figsize =(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=df['diagnosis'])
plt.xlabel('1st principle component',fontsize=15)
plt.ylabel('2nd principle component',fontsize=15)
plt.title("PCA Graph",fontsize=15)
```

```python
x = feature_df.values
x = StandardScaler().fit_transform(x)

pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```
