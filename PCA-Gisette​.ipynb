```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy.random import randint
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('gisette_train.data', header=None)
df.head()
#df1.shape
```

```python
df[0] = df[0].map(lambda x: x.strip())
df2 = df[0].str.split(" ",expand=True)
df2
```

```python
feature_df = df2.iloc[:,:]
feature_df
```

```python
label_df = df2.iloc[:,0]
label_df
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

x = df2.values
x = StandardScaler().fit_transform(x)
np.mean(x),np.std(x)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df2)
scaled_data = scaler.transform(df2)

from sklearn.decomposition import PCA
pca  = PCA(n_components=2)
pc = pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data);
scaled_data.shape
```

```python
x_pca.shape

plt.figure(figsize =(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1])
plt.xlabel('1st principle component',fontsize=15)
plt.ylabel('2nd principle component',fontsize=15)
plt.title("PCA Graph",fontsize=15)
    
```

```python
x = feature_df.values
x = StandardScaler().fit_transform(x)

pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```
